```python
# Import necessary module
import os
import pandas
import boto3 
from dotenv import load_dotenv
```

```python
# 
load_dotenv()

AWS_ACCESS_KEY = os.getenv('AWS_ACCESS_KEY')
AWS_SECRET_KEY = os.getenv('AWS_SECRET_KEY')
REGION = os.getenv('REGION')
AWS_BUCKET = os.getenv('AWS_BUCKET')
OUTPUT_FORTMAT = os.getenv('OUTPUT_FORTMAT')

s3 = boto3.client(
            "s3",
            aws_access_key_id=AWS_ACCESS_KEY,
            aws_secret_access_key=AWS_SECRET_KEY,
            region_name=REGION
            )

s3.upload_file(
    Filename= 'data/weather_data.csv',
    Bucket= AWS_BUCKET,
    Key= 'raw/weather_data.csv'
)

print('File Uploaded Successfully')
```

```python
# Create a function to load data into S3 bucket
def fn_load_csv_s3b(filename='data/weather_data.csv', bucket=None, key=None, aws_access_key=None, aws_secret_key=None, region=None):
    """Upload a local CSV file to S3.
    Parameters:
    - filename: local path to the file
    - bucket: S3 bucket name (defaults to AWS_BUCKET from env)
    - key: S3 object key (defaults to 'raw/<basename>')
    - aws_access_key, aws_secret_key, region: AWS credentials (default to env values)
    Returns True on success, False on failure."""
    # Use environment values defined earlier in the notebook if no explicit values passed
    bucket = bucket or AWS_BUCKET
    aws_access_key = aws_access_key or AWS_ACCESS_KEY
    aws_secret_key = aws_secret_key or AWS_SECRET_KEY
    region = region or REGION
    if key is None:
        key = os.path.basename(filename)
    s3_client = boto3.client('s3', aws_access_key_id=aws_access_key, aws_secret_access_key=aws_secret_key, region_name=region)
    try:
        s3_client.upload_file(Filename=filename, Bucket=bucket, Key=key)
        print(f'File {filename} uploaded to s3://{bucket}/{key}')
        return True
    except Exception as e:
        print('Upload failed:', e)
        return False

# Example usage (uncomment to run):
fn_load_csv_s3b()
# fn_load_csv_s3b(filename='data/weather_data.csv', key='raw/weather_data.csv')
```
